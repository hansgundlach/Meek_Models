```python
#now model the benchmark performanc
benchmark_func = lambda x: np.log(x) + 1
#lamb

# performance_at_param_time = lambda (param, time): np.log(param**2)+np.log(param)
g_train = 2
g_inf = 5
def performnace_at_given_time(time, cost):
    train_available = (cost**2*(g_train*g_inf)**time)
    inf_available = (cost*(g_inf)**time)
    perform = np.log(train_available) + np.log(inf_available)
    return perform


# now we can compare performanc at cost over time 

times = np.linspace(0, 20, 100)
perform_10 = [performnace_at_given_time(t, 10) for t in times]
perform_100 = [performnace_at_given_time(t, 100) for t in times]
perform_10000 = [performnace_at_given_time(t, 10000) for t in times]

import matplotlib.pyplot as plt
plt.plot(times, perform_10, label="cost = 10")
plt.plot(times, perform_100, label="cost = 100")
plt.plot(times, perform_10000, label="cost = 10000")
plt.legend()
plt.show()

```
