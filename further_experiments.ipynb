import matplotlib as plt

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# comparison of loss difference and token differentation threshold 
discrepancy_factor = 2.65

token_threshold = lambda loss_diff: discrepancy_factor/loss_diff
loss_diffs = np.logspace(-5, 1, 100)
plt.figure(figsize=(10, 6))
plt.ylim(1e0, 1e5)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.plot(loss_diffs, token_threshold(loss_diffs), label='Token Threshold (Slowdown Factor 1)', linewidth=3, alpha=0.99)
plt.plot(loss_diffs,10*token_threshold(loss_diffs), label='Token Threshold (Slowdown Factor 10)', linewidth=3, alpha=0.99)
plt.plot(loss_diffs,100*token_threshold(loss_diffs), label='Token Threshold (Slowdown Factor 100)', linewidth=3, alpha=0.99)
plt.xlabel('Loss Difference', fontsize=20)
plt.ylabel('Token Threshold', fontsize=20)
# plt.title('Token Threshold vs Loss Difference', fontsize=14, fontweight='bold')
plt.yticks(fontsize=20)
plt.xticks(fontsize=20)
plt.yscale('log')
plt.xscale('log')
plt.axhline(y=1e3, color='black', linestyle='--', linewidth=2)
plt.text(10**(-5.2), 1.2e3, 'length of short blog post', fontsize=20)

plt.axhline(y=1e2, color='black', linestyle='--', linewidth=2)
plt.text(10**(-5.2), 1.2e2, 'tweet-length', fontsize=20)
plt.legend(fontsize=18)
# plt.grid(True)
plt.tight_layout()
plt.savefig("Figures/token_threshold_vs_loss_difference.png")
plt.show()
```
